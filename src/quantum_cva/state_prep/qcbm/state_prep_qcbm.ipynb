```python
import numpy as np
from quantum_cva.state_prep.qcbm.p_target import build_joint_target_from_P_bin

# Toy target: M=2 times, N=2 price-bins (no padding needed)
P_bin = np.array([
    [0.25, 0.75],  # P(s=0|t=0), P(s=1|t=0)
    [0.60, 0.40],  # P(s=0|t=1), P(s=1|t=1)
], dtype=float)

tgt = build_joint_target_from_P_bin(P_bin, order="time_major", pad_to_pow2=True)

# With uniform P(t)=1/2, the joint is p(t,s)=0.5*P(s|t)
expected = np.array([
    0.5 * 0.25,  # (t=0, s=0) -> |00>
    0.5 * 0.75,  # (t=0, s=1) -> |01>
    0.5 * 0.60,  # (t=1, s=0) -> |10>
    0.5 * 0.40,  # (t=1, s=1) -> |11>
], dtype=float)

assert tgt.M_pad == 2 and tgt.N_pad == 2
assert tgt.m == 1 and tgt.n == 1
assert np.allclose(tgt.p_tg, expected, atol=1e-12)
assert abs(tgt.p_tg.sum() - 1.0) < 1e-12

print("p_tg =", tgt.p_tg)

# Interpret as computational basis probabilities for |t s> (time bit then price bit)
labels = ["|00> (t0,s0)", "|01> (t0,s1)", "|10> (t1,s0)", "|11> (t1,s1)"]
for lab, p in zip(labels, tgt.p_tg):
    print(f"{lab}: {p:.6f}")
```
