```python
from qiskit import QuantumCircuit
from qiskit.circuit import ParameterVector

def build_ansatz_fig5():
    n_qubits = 4
    n_params = 14
    theta = ParameterVector("Оё", n_params)

    qc = QuantumCircuit(n_qubits, name="G_p_fig5")

    # l = 1 (single-qubit layer): Rx, Rz on each qubit
    k = 0
    for q in range(n_qubits):
        qc.rx(theta[k], q); k += 1
        qc.rz(theta[k], q); k += 1

    # l = 2 (entangling layer): 6 XX gates (use RXX in Qiskit)
    pairs = [(0, 1), (1, 2), (2, 3), (0, 1), (1, 2), (2, 3)]
    for (a, b) in pairs:
        qc.rxx(theta[k], a, b); k += 1

    assert k == n_params
    return qc, theta

qc, theta = build_ansatz_fig5()
print(qc)
print("n_params =", len(theta))
```
